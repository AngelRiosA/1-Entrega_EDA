- Tendencia general: Todos los países muestran una tendencia a la baja en las tasas de abandono escolar a lo largo del período.

- Alemania y Francia: Ambos países muestran las tasas más bajas y estables del grupo. Francia logra una reducción gradual, llegando a 7.8% en 2023, mientras que Alemania fluctúa ligeramente, terminando en 8.7%.
